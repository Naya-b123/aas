```python
# Input
distance = 0.35   # meters

# Initial velocity
velocity = 5      # assume robot moving at 5 m/s

# Obstacle detection logic
if distance < 0.5:
    velocity = 0
    status = "Robot Halted Safely"
else:
    status = "Robot Moving"

# Output
print("Status:", status)
print("Velocity:", velocity)
```
