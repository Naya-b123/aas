```python
# Simulated key input
key = 'A'

# Teleoperation mapping
if key.upper() == 'A':
    action = "Left Turn"
elif key.upper() == 'D':
    action = "Right Turn"
elif key.upper() == 'W':
    action = "Move Forward"
elif key.upper() == 'S':
    action = "Move Backward"
else:
    action = "No Action"

# Output
print("Action:", action)
```
